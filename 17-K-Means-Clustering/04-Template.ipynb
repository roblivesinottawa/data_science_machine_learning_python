# K MEANS CLUSTERING

kmeans clustering is a simple and effective method for cluster analysis. it is an unsupervised learning algorithm whicich tries to cluster data based on their similarity.


```python
# import visulization libraries
import seaborn as sns   
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# CREATE A DATASET WITH ARTIFICIAL DATA
from sklearn.datasets import make_blobs

data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data[0]
```

```python
# VISUALIZE THE DATASET
plt.figure(figsize=(12,8))
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
# CREATING THE CLUSTERS
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')
ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```
